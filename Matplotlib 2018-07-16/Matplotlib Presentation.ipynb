## Single-Plot Figures
### Scatter Plot
Starting with the basics.

First, we'll create some numpy arrays to use as input for the plots.

```python
# Create some data for plotting
x = np.arange(1,11)
y = np.arange(2,22,step=2)
z = np.arange(20,0,step=-2)
x_labels = np.arange(10,110, step=10)
hist_normal = np.random.normal(5,2,30000)
```

This is the smallest amount of code needed to create a scatter plot figure in matplotlib. 

```python
# create a figure
plt.figure() # creates a figure object
plt.scatter(x, y) # adds a scatter plot (one axes) to your figure object
plt.show() # displays your plot
```

This plot is pretty bare-bones. It shows your data, but there is definitely more you can do to make it look better. Once you have a skeleton for your scatter plot set up, you can then start customising your plot to make it more informative and visually appealing. 
Let's start with `plt.figure()`. Adding `figsize=(width,height)` as a parameter changes the size of the figure to width x height in inches. `dpi=` will change the dpi to any integer number you choose, and `facecolor=` will change the face of the figure (note, *not* the axes) to the colour you specify. 
Let's see how that changes the figure.

```python
# create a figure
plt.figure(figsize=(8,5), dpi=100, facecolor='pink')
plt.scatter(x, y)
plt.show()
```

I'm not much of a fan of the pink face colour, but it gives you a good idea of what `facecolor=` does. 
Now let's have a look at how we can customise the scatter plot. We can change the colour of the markers using `c=`, the marker shape using `marker=`, the marker size using `s=`, and the transparency using `alpha=`. I'm also going to remove the pink face colour to make the plot a bit more visually appealing. 

```python
# create a figure
plt.figure(figsize=(8,5), dpi=100)
plt.scatter(x, y, c='orange', marker='D', alpha=0.7, s=50)
plt.show()
```

Next, we're going to want to label our graph and axes using `plt.title()`, `plt.xlabel()` and `plt.ylabel()`. Each of these methods take similar arguments, such as a string for the label, font size, and colour. 

```python
# create a figure
plt.figure(figsize=(8,5), dpi=100)
plt.scatter(x, y, c='orange', marker='D', alpha=0.7, s=50)
plt.title('Scatter Plot', size=20, color='blue')
plt.xlabel('x-axis', size=15)
plt.ylabel('y-axis', size=15)
plt.show()
```

This is starting to look a bit better, but there's still more we can do to customise the plot. We can change the labels on the axes (ie. the x-axis and y-axis) using `plt.xticks()` and `plt.yticks()`. They both take a list or array of the tick locations you want to label, and the labels you want to use. 

```python
# create a figure
plt.figure(figsize=(8,5), dpi=100)
plt.scatter(x, y, c='orange', marker='D', alpha=0.7, s=50)
plt.title('Scatter Plot', size=20, color='blue')
plt.xlabel('x-axis', size=15)
plt.ylabel('y-axis', size=15)
plt.show()
```

You may, for some reason, want to show multiple data sets in the one plot. You can do this by adding another line of `plt.scatter()` (or any other plot, such as `plt.plot()`) below the current one. I'll add two more using the `y` and `z` arrays I created earlier.

```python
# create a figure
plt.figure(figsize=(8,5), dpi=100)
plt.scatter(x, y, c='orange', marker='D', alpha=0.7, s=50)
plt.scatter(y, x, c='blue', marker='x', alpha=0.7)
plt.plot(y, z, c='red', marker='*', alpha=0.5)
plt.title('Scatter Plot', size=20, color='blue')
plt.xlabel('x-axis', size=15)
plt.ylabel('y-axis', size=15)
plt.show()
```

Now we're getting somewhere. Now that we've got multiple data representations in one figure, we're going to want to add a legend. To do that we add `label=` to `plt.scatter()`, and a new line with `plt.legend()`, which will automatically import the label names that we've just specified.

```python
# create a figure
plt.figure(figsize=(8,5), dpi=100)
plt.scatter(x, y, c='orange', marker='D', alpha=0.7, s=50, label='diamonds')
plt.scatter(y, x, c='blue', marker='x', alpha=0.7, label='Xs')
plt.plot(y, z, c='red', marker='*', alpha=0.5, label='stars')
plt.legend()
plt.title('Scatter Plot', size=20, color='blue')
plt.xlabel('x-axis', size=15)
plt.ylabel('y-axis', size=15)
plt.show()
```

Finally, we're going to change the x-axis labels. To do this we will use `plt.xticks()`, and give it an array of x-axis locations, and the labels we want at those locations. We can also rotate the labels using `rotation=`.

```python
# create a figure
plt.figure(figsize=(8,5), dpi=100)
plt.scatter(x, y, c='orange', marker='D', alpha=0.7, s=50, label='diamonds')
plt.scatter(y, x, c='blue', marker='x', alpha=0.7, label='Xs')
plt.plot(y, z, c='red', marker='*', alpha=0.5, label='stars')
plt.legend()
plt.title('Scatter Plot', size=20, color='blue')
plt.xlabel('x-axis', size=15)
plt.ylabel('y-axis', size=15)
plt.xticks(np.arange(0,21,step=5), ['zero', 'five', 'ten', 'fifteen', 'twenty'], rotation=40)
plt.show()
```

There is still a lot more we can do with this plot. But for now, we're going to move on to the next plot. If you're interested, more detailed information for scatter plot creation can be found at https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html.

### Heatmaps using imshow()

Imshow can be used for a lot more than just heatmaps, but that's what I'll focus on for this tutorial. All of the code apart from `imshow()` is the same as before, so I'll pre-fill some to save some time.

```python
# imshow can take 2D numpy arrays and pandas dataframes
data = np.random.normal(0, 2, size=(50,50))
df = pd.DataFrame(data)

# Plotting a heatmap figure
plt.figure(figsize=(5,5), dpi=100)
plt.imshow(df)
plt.title('Heatmap', size=20)
plt.xlabel('x-axis', size=15)
plt.ylabel('y-axis', size=15)
plt.show()
```

The biggest change to make to `imshow()` is to change the colour map using `cmap=`. Take a look here to get a better idea of what different colour maps are available: https://matplotlib.org/users/colormaps.html
I'm going to use magma for this, but there are many other colour maps to choose from.

```python
# imshow can take 2D numpy arrays and pandas dataframes
data = np.random.normal(0, 2, size=(50,50))
df = pd.DataFrame(data)

# Plotting a heatmap figure
plt.figure(figsize=(5,5), dpi=100)
plt.imshow(df, cmap='magma')
plt.title('Heatmap', size=20)
plt.xlabel('x-axis', size=15)
plt.ylabel('y-axis', size=15)
plt.show()
```

That's alredy looking pretty good, but we need to know what the colours mean. To create a colour bar, use `plt.colorbar()`, which will take the colours and values from your heatmap.

```python
# imshow can take 2D numpy arrays and pandas dataframes
data = np.random.normal(0, 2, size=(50,50))
df = pd.DataFrame(data)

# Plotting a heatmap figure
plt.figure(figsize=(5,5), dpi=100)
plt.imshow(df, cmap='magma')
plt.colorbar()
plt.title('Heatmap', size=20)
plt.xlabel('x-axis', size=15)
plt.ylabel('y-axis', size=15)
plt.show()
```

That shows us a colour bar, but it's not aligned very nicely. The simple solution to that is to add the argument `aspect='auto'` to `plt.imshow()`, which will snap everything to the same grid.

```python
# imshow can take 2D numpy arrays and pandas dataframes
data = np.random.normal(0, 2, size=(50,50))
df = pd.DataFrame(data)

# Plotting a heatmap figure
plt.figure(figsize=(5,5), dpi=100)
plt.imshow(df, cmap='magma', aspect='auto')
plt.colorbar()
plt.title('Heatmap', size=20)
plt.xlabel('x-axis', size=15)
plt.ylabel('y-axis', size=15)
plt.show()
```

That's a lot better. Finally, I'm going to get rid of the line around the colourbar, by naming the colourbar object, and using the `outline.set_linewidth()` method.

```python
# imshow can take 2D numpy arrays and pandas dataframes
data = np.random.normal(0, 2, size=(50,50))
df = pd.DataFrame(data)

# Plotting a heatmap figure
plt.figure(figsize=(5,5), dpi=100)
plt.imshow(df, cmap='magma', aspect='auto')
cb = plt.colorbar()
cb.outline.set_linewidth(0)
plt.title('Heatmap', size=20)
plt.xlabel('x-axis', size=15)
plt.ylabel('y-axis', size=15)
plt.show()
```

You can find a lot more information on `imshow()` here: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.imshow.html

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,5), dpi=100)

# Leftmost subplot
ax1.scatter(x, y)

# Rightmost subplot
f2 = ax2.imshow(df)

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,5), dpi=100)

# Leftmost subplot
ax1.scatter(x, y, c='coral', marker='D', alpha=0.7, s=50, label='diamonds')
ax1.scatter(y, x, c='blue', marker='x', alpha=0.7, label='Xs')
ax1.scatter(y, z, c='red', marker='*', alpha=0.5, label='stars')
ax1.legend()
ax1.set_title('Subplot1 - Scatter Plot', size=20)
ax1.set_xlabel('X1')
ax1.set_ylabel('Y1')

# Rightmost subplot
f2 = ax2.imshow(df, cmap='magma', aspect='auto')
ax2.set_title('Subplot2 - Heatmap', size=20)
ax2.set_xlabel('X2', size=15)
ax2.set_ylabel('Y2', size=15)
cb = fig.colorbar(f2,ax=ax2)
cb.outline.set_linewidth(0)

plt.show()
```

## Multi-Plot Figures with Different Dimensions 
### Using subplot2grid()
For creating custom plots that don't necessarily align to one square in a grid.

```python
# Setting up the grid, and each plot within the grid

gridsize = (3,2)
fig = plt.figure(figsize=(10,12))
ax1 = plt.subplot2grid(gridsize, (0, 0), colspan=2, rowspan=2)
ax2 = plt.subplot2grid(gridsize, (2,0)) # colspan & rowspan are 0 by default
ax3 = plt.subplot2grid(gridsize, (2,1))
```

## Filling in the Plots
Once the axes have been assigned, filling them in using `plt.subplot2grid()` is similar to what we did using `plt.subplots()`.

```python
gridsize = (3,2)
fig = plt.figure(figsize=(10,12))

ax1 = plt.subplot2grid(gridsize, (0, 0), colspan=2, rowspan=2)
heatmap = ax1.imshow(df, cmap='magma', aspect='auto')
ax1.set_title('Heatmap', size=15)
cb = plt.colorbar(heatmap, ax=ax1)
cb.outline.set_linewidth(0)

ax2 = plt.subplot2grid(gridsize, (2,0)) # colspan & rowspan are 0 by default
ax2.scatter(x, y, c='coral', marker='D', alpha=0.7, s=50)
ax2.scatter(y, x, c='blue', marker='x', alpha=0.7)
ax2.scatter(y, z, c='red', marker='*', alpha=0.5)
ax2.set_title('Scatter', size=15)

ax3 = plt.subplot2grid(gridsize, (2,1))
ax3.hist(hist_normal, color='skyblue')
ax3.set_title('Histogram', size=15)
```

### Clean axis function
For my figures I like to remove the default axes (ie. plural of axis, not the whole plot) and build what I want from scratch. Calling the clean_axis() function is an easy way to do that when you're working with a lot of plots in one figure.

```python
def clean_axis(ax):
    '''Remove ticks, tick labels, and frame from axis'''
    ax.get_xaxis().set_ticks([])
    ax.get_yaxis().set_ticks([])
    for sp in ax.spines.values():
        sp.set_visible(False)
```

```python
# Setting up the figure
fig = plt.figure(figsize=(6,5), dpi=150) 

# Creating a custom grid using gridspec
heatmapGS = gridspec.GridSpec(2, 2, wspace=0.0, hspace=0.0, width_ratios=[1, 0.15], height_ratios=[0.05, 1.5])

# Heatmap
heatmapAX = fig.add_subplot(heatmapGS[1, 0])
axi = heatmapAX.imshow(rnaseq_df, cmap='seismic', vmin=-3.5,
                       vmax=3.5, aspect='auto', origin='lower')
clean_axis(heatmapAX)
heatmapAX.yaxis.set_ticks_position('left')
heatmapAX.set_yticks(np.arange(rnaseq_df.shape[0]))
heatmapAX.set_yticklabels(gene_names)
plt.yticks(size=8)

# Colorbar
cbGS = gridspec.GridSpecFromSubplotSpec(1, 3, subplot_spec=heatmapGS[1, 1], wspace=0.0, hspace=0.1)
cbAX = fig.add_subplot(cbGS[0, 1])
cbAX.set_title('z-score', size=8)
cb = fig.colorbar(axi, cbAX) 
cb.outline.set_linewidth(0)
ticks = cb.ax.yaxis.get_ticklabels()
for tick in ticks:
    tick.set_fontsize(tick.get_fontsize() - 2)

# Colours for column colourbar
list_cbar = []
low, med, high = 100, 100, 100
immune_expression = [low, med, high]
for i in range(3): #low, medium and high immune expression
    expression = immune_expression[i]
    loc_cmap = [brewer2mpl.get_map('Spectral', 'Diverging', 3).mpl_colors[i]] * expression
    list_cbar += loc_cmap

# add column colorbar
col_cbAX = fig.add_subplot(heatmapGS[0, 0])
col_axi = col_cbAX.imshow([list_cbar], aspect='auto', shape=(len(list_cbar), 1))
clean_axis(col_cbAX)
col_cbAX.xaxis.set_ticks_position('top')
col_cbAX.set_xticks([50, 150, 250])
col_cbAX.set_xticklabels(['Low', 'Med', 'High'])


fig.tight_layout()
plt.show()
```
