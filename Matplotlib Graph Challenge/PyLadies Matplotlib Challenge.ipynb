# PyLadies Matplotlib Graph Challenge

```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# import data into a pandas dataframe
members_df = pd.read_table('PyLadies_Melbourne_Total_and_Active_Members.xls')
members_df.head()
```

```python
# Remove white space from the outside of column names
col_names = [col_name.strip() for col_name in members_df.columns]
members_df.columns = col_names

# change 'Date' column to the index
members_df.set_index('Date', drop=True, inplace=True)

members_df.head()
```

```python
# get dates only for the first day of each month
date_list = members_df.index.tolist()
dates = [date for date in date_list if date.strip()[-2:] == '01']
plot_dates = [date_list[0]] + dates
plot_dates
```

```python
# select members only for plot_dates
plot_df = members_df.loc[plot_dates]
plot_df.head()
```

```python
# create a list for each column
memb_total = plot_df['Total Members'].tolist()
memb_active = plot_df['Active Members'].tolist()
```

```python
# create the plot
fig, ax = plt.subplots(figsize=(11, 7), dpi=200)

# active members
ax.stackplot(plot_dates, memb_active, color='red', alpha=0.3)
ax.plot(plot_dates, memb_active, c='red', marker='o', label='Active Members')

# total members
ax.stackplot(plot_dates, memb_total, color='royalblue', alpha=0.3)
ax.plot(plot_dates, memb_total, c='royalblue', marker='o', label='Members')

# remove white space from around the inside of the plot
ax.set_xlim(plot_dates[0], plot_dates[-1])

# plot customisation
plt.title('Total and Active Members', size=25, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.yticks(np.arange(0, 301, step=50))
plt.grid(which='major', axis='y')
plt.legend(loc='lower center', ncol=2, bbox_to_anchor=(0.5, -0.25), fontsize='large', frameon=False)

plt.show()
```
